```python
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from pathlib import Path

# Load image paths
real_images = list(Path('/content/drive/MyDrive/data/real').glob('*'))
spoof_images = list(Path('/content/drive/MyDrive/data/fake').glob('*'))

# Create labels
real_labels = ['1'] * len(real_images)
spoof_labels = ['0'] * len(spoof_images)

# Create a DataFrame with image paths and labelsQ
df = pd.DataFrame({'filename': [str(img) for img in real_images + spoof_images],
                   'label': real_labels + spoof_labels})

# Verify DataFrame
print(df.head())

# Split into training and validation sets
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)

# Data augmentation for training set
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Data generator for validation set (no augmentation)
val_datagen = ImageDataGenerator(rescale=1./255)

# Generate training data from DataFrame
train_generator = train_datagen.flow_from_dataframe(
    dataframe=train_df,
    x_col='filename',
    y_col='label',
    target_size=(160, 160),
    batch_size=32,
    class_mode='binary'
)

# Generate validation data from DataFrame
val_generator = val_datagen.flow_from_dataframe(
    dataframe=val_df,
    x_col='filename',
    y_col='label',
    target_size=(160, 160),
    batch_size=32,
    class_mode='binary'
)

# Check the number of images found
print(f"Found {train_generator.samples} training images.")
print(f"Found {val_generator.samples} validation images.")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming class_names are defined
class_names = ['fake','real']

# Extract a batch of images and labels from the generator
train_images, train_labels = next(train_generator)

# Number of images to display
num_images = 315
rows = 15
cols = 21

plt.figure(figsize=(20, 15))  # Adjust figsize to fit the grid

for i in range(min(num_images, len(train_images))):
    plt.subplot(rows, cols, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i])

    # Convert the label to integer and map to class names
    plt.xlabel(class_names[int(train_labels[i])])

plt.tight_layout()  # Adjust subplots to fit into the figure area
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Load the MobileNetV2 model pre-trained on ImageNet, excluding the top layer
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(160, 160, 3))

# Freeze the base model layers to prevent them from being trained
base_model.trainable = False

# Add custom layers on top of the base model
x = base_model.output
x = GlobalAveragePooling2D()(x)  # Global average pooling layer
x = Dense(128, activation='relu')(x)  # Fully connected layer
x = Dropout(0.5)(x)  # Dropout for regularization
output = Dense(1, activation='sigmoid')(x)  # Output layer for binary classification

# Create the final model
model = Model(inputs=base_model.input, outputs=output)

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model, Model, model_from_json
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.backend import clear_session

# Load model architecture and weights
model_path = 'best_liveness_model_mobilenetv2.keras'
if os.path.exists(model_path):
    print("Loading pre-trained model...")
    model = load_model(model_path)
else:
    print("Pre-trained model not found. Please ensure you've saved a checkpoint model first.")
    # Optionally, you could train the model here if it does not exist.
    exit()

# Print model summary to check input/output shapes
model.summary()

# Function to preprocess the image
def preprocess_image(image_path, target_size=(160, 160)):
    img = load_img(image_path, target_size=target_size)
    img_array = img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = preprocess_input(img_array)
    return img_array

# Function to make predictions
def predict_image(image_path):
    processed_image = preprocess_image(image_path)
    prediction = model.predict(processed_image)
    if prediction[0][0] > 0.5:
        return "Real", prediction[0][0]
    else:
        return "Spoof", prediction[0][0]

# Path to the folder containing test images
test_images_folder = "/content/drive/MyDrive/test_images"  # Replace with the path to your test images folder

# Get a list of all images in the folder
test_image_paths = [os.path.join(test_images_folder, img) for img in os.listdir(test_images_folder) if img.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Loop through each image, make predictions, and display results
for image_path in test_image_paths:
    prediction_label, prediction_score = predict_image(image_path)
    print(f"Image: {os.path.basename(image_path)} | Prediction: {prediction_label} | Score: {prediction_score:.2f}")

    # Display the image with its prediction
    img = load_img(image_path)
    plt.imshow(img)
    plt.title(f'{os.path.basename(image_path)}\nPrediction: {prediction_label} | Score: {prediction_score:.2f}')
    plt.axis('off')
    plt.show()
```

```python
def update_input_shape(config):
    if 'batch_shape' in config:
        config['batch_input_shape'] = config.pop('batch_shape')
    return config

def update_model_json(model_json):
    if 'config' in model_json:
        model_json['config'] = update_input_shape(model_json['config'])
        if 'layers' in model_json['config']:
            for layer in model_json['config']['layers']:
                if 'config' in layer:
                    layer['config'] = update_input_shape(layer['config'])
    return model_json

updated_model_json = update_model_json(model_json)

# Save the updated JSON file
with open('/content/drive/MyDrive/tfjs_model/model_updated.json', 'w') as file:
    json.dump(updated_model_json, file, indent=2)
```
